```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
data=pd.read_excel('学生A工作打分.xlsx')
datamain=data.values.tolist()
 
'针对对特征Parameter,利用SSE选择最优k'
SSE = []  # 存放每次结果的误差平方和
for k in range(1, 9):  # K的范围 ： 1-9
    estimator = KMeans(n_clusters=k)  # 构造聚类器
    estimator.fit(data[['A']])
    SSE.append(estimator.inertia_)
X = range(1, 9)
plt.xlabel('k')
plt.ylabel('SSE')
plt.plot(X, SSE, 'o-')
plt.show()
```

```python

dataK=np.array(data) #数据矩阵化
estimator = KMeans(n_clusters=3) #Kmeans提取特征 聚类为3类
res = estimator.fit_predict(data)
lable_pred = estimator.labels_
centroids = estimator.cluster_centers_
inertia = estimator.inertia_
 
print("lable_pred:", lable_pred) #分类结果
print("centroids:", centroids) #聚类中心
print("inertia:", inertia) #内点
 
for i in range(len(dataK)):
    if int(lable_pred[i]) == 0:
        plt.scatter(dataK[i][0],dataK[i][1], color='black')
    if int(lable_pred[i]) == 1:
        plt.scatter(dataK[i][0],dataK[i][1], color='red')
    if int(lable_pred[i]) == 2:
        plt.scatter(dataK[i][0],dataK[i][1], color='blue')

plt.show()
```
